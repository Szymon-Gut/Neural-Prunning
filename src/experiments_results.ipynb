```python
import matplotlib.pyplot as plt

pruning_levels = [0, 30, 40, 50, 70, 80, 90, 95]
accuracies = [87, 86.5, 86.4, 85, 82, 76, 63, 49]

best_compromise_index = pruning_levels.index(50)
best_pruning = pruning_levels[best_compromise_index]
best_accuracy = accuracies[best_compromise_index]

plt.figure(figsize=(10, 6))
plt.plot(pruning_levels, accuracies, marker='o', linestyle='-', color='royalblue', label='Accuracy')
plt.axvline(x=best_pruning, color='red', linestyle='--', label=f'Najlepszy kompromis ({best_pruning}%)')
plt.scatter(best_pruning, best_accuracy, color='red', s=80, zorder=5)

plt.title('Wpyw stopnia prunningu na dokadno modelu', fontsize=14)
plt.xlabel('Stopie prunningu (%)', fontsize=12)
plt.ylabel('Dokadno na zbiorze testowym (%)', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
